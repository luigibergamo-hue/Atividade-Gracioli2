```
# EXERCÍCIO 18: Conte vogais
print("## EXERCÍCIO 18 ##")
print("## Conte vogais ##")
def contar_vogais(texto):
    vogais = "aeiouAEIOU"
    contador = 0
    for letra in texto:
        if letra in vogais:
            contador += 1
    return contador

frase = "Python é legal"
print(contar_vogais(frase))
```

```
# Exercício 19: Conte palavras de uma frase
print("## Exercício 19 ##")
print("## Conte palavras de uma frase ##")
texto = "Python é divertido e versátil"
palavras = texto.split()
total = len(palavras)
print(f"Total de palavras: {total}")
```

```
# Exercício 20: Função que soma dois números
print("## Exercício 20 ##")
print("## Função que soma dois números ##")
def somar(a, b):
  return a + b
resultado = somar(5, 3)
print(resultado)
```

```
# Exercício 21: Função que retorna maior número
print("## Exercício 21 ##")
print("## Função que retorna maior número ##")
numero = [10, 5, 8, 22, 3]
maior = max(numero)
print(maior)
```

```
# Exercício 22: Função que verifica par
print("## Exercício 22 ##")
print("## Função que verifica par ##")
def eh_par(numero):
  """Verifica se um número é par."""
  return numero % 2 == 0
numero = 10
if eh_par(numero):
    print(f"{numero} é par.")
else:
    print(f"{numero} é ímpar.")
```
