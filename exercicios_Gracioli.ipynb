## 1
Leia um número inteiro e exiba o seu sucessor e antecessor.

```
numero = int(input("Digite um número inteiro: "))

sucessor = numero + 1
print("O sucessor é:", sucessor)

antecessor = numero - 1
print("O antecessor é:", antecessor)
```

## 2
Leia dois números e apresente a soma, subtração, multiplicação e divisão entre eles.

```
numero1 = int(input("Digite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))

soma = numero1 + numero2
print(numero1, "+", numero2, "=", soma)

subtracao = numero1 - numero2
print(numero1, "-", numero2, "=", subtracao)

multiplicacao = numero1 * numero2
print(numero1, "*", numero2, "=", multiplicacao)

divisao = numero1 / numero2
print(numero1, "/", numero2, "=", divisao)
```

## 3
Leia a idade de uma pessoa em anos e converta para meses e dias.

```
idade = int(input("Digite a sua idade: "))

meses = idade * 12
dias = idade * 365

print(f"Idade em anos: {idade}")
print(f"Aproximadamente {meses} meses.")
print(f"Aproximadamente {dias} dias.")
```

## 4
Leia um número e verifique se ele é positivo, negativo ou zero.

```
numero = float(input("digite um número: "))
if numero >= 1:
  print("O número é positivo!")
elif numero <= 0:
  print("O número é negativo!")
else:
  print("O número é zero!")
```

## 5
Leia um número inteiro e verifique se ele é par ou ímpar.

```
numero = int(input("Digite um número: "))
if numero % 2 == 0:
  print("O número é par!")
else:
  print("O número é ímpar!")
```

## 6
Leia três números e determine qual é o maior entre eles.

```
numero1 = int(input("Digite o primeiro número: "))
numero2 = int(input("Digite o segundo número: "))
numero3 = int(input("Digite o terceiro número: "))

maior = max(numero1, numero2, numero3)
print(f"O maior número é: {maior}")
```

## 7
Leia três notas de um aluno, calcule a média e informe se ele foi aprovado, reprovado ou está em recuperação.

```
nota1 = float(input("Digite a primeira nota: "))
nota2 = float(input("Digite a segundo nota:"))
nota3 = float(input("digiye a terceira nota: "))

media = int(nota1 + nota2 + nota3) / 3
print(f"Média: {media:.2f}")

if media >= 7.0:
    print("Situação: Aprovado")
elif media >= 5.0:
    print("Situação: Recuperação")
else:
    print("Situação: Reprovado")
```

## 8
Leia um número e calcule o seu fatorial.

```
import math

numero = int(input("Digite um número: "))

numero = int(numero)
resultado = math.factorial(numero)
print(f"O fatorial e {numero} é {resultado}")
```

## 9
Leia um número e exiba a tabuada desse número de 1 a 10.

```
numero = int(input("Digite um número para ver a tabuada"))
print(f"tabuada de {numero}: ")
for i in range(1,11):
  print(f"{numero} x {i} = {numero * i}")
```
