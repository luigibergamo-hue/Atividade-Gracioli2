# TREINO DE PYTHON

```
alunos = []

while True:

 print("## ========== MENU ========== ##")
 print("## 1 - Cadastrar aluno ##")
 print("## 2 - Lançar notas ##")
 print("## 3 - Ver Boletim ##")
 print("## 4 - Média da turma ##")
 print("## 5 - Situação da turma ##")
 print("## 6 - Remover aluno ##")
 print("## 7 - Listar alunos ##")
 print("## 0 - Sair ##")

 escolha = int(input("Escolha: "))

 match escolha:
  case 1:
    # Cadastrar_aluno()
    nome = input("Digite o nome do aluno: ")
    aluno = {
        "nome": nome,
        "notas": []
    }

    alunos.append(aluno)
    print("Aluno cadastrado com sucesso!")

  case 2:
    # Lançar_notas()
    nome = input("Digite o nome do aluno: ")
    encontrado = False

    for aluno in alunos:
      if aluno["nome"] == nome:
        encontrado = True

        qtd = int(input("Quantas notas você deseja lançar? "))

        for i in range(qtd):
          nota = float(input(f"Digite a nota {i+1}: "))
          aluno["notas"].append(nota)

        print("notas lançadas com sucesso!")
        break
    if not encontrado:
      print("Aluno não encontrado.")

  case 3:
    # Ver_Boletim()
    nome = input("Digite o nome do aluno: ")
    encontrado = False

    for aluno in alunos:
      if aluno["nome"] == nome:
        encontrado = True
        print(f"\nBoletim de {aluno["nome"]}")

        if len(aluno["notas"]) == 0:
          print("Nenhuma nota cadastrada.")
        else:
          for i, nota in enumerate(aluno["notas"]):
            print(f"Nota {i+1}: {nota}")

          media = sum(aluno["notas"]) / len(aluno["notas"])
          print(f"Média: {media:.2f}")
        break

    if not encontrado:
      print("Aluno não encontrado.")

  case 4:
   # Média_da_turma()
    todas_notas = []

    for aluno in alunos:
      todas_notas.extend(aluno["notas"])

    if len(todas_notas) == 0:
      print("Nenhuma nota cadastrada na turma.")
    else:
      media_turma = sum(todas_notas) / len(todas_notas)
      print(f"Média da turma: {media_turma:.2f}")

  case 5:
    # Situacao_da_turma()
    if len(alunos) == 0:
      print("Nenhum aluno cadastrado.")
    else:
      for aluno in alunos:
        print(f"\nAluno: {aluno["nome"]}")

        if len(aluno["notas"]) == 0:
          print("Sem notas.")
        else:
          media = sum(aluno["notas"]) / len(aluno["notas"])
          print(f"Média: {media:.2f}")

          if media >= 7:
            print("Situação: Aprovado.")
          elif media >= 5:
            print("Situação: Recuperação.")
          else:
            print("Situação: Reprovado.")
  case 6:
    # remover_aluno()
    nome = input("Digite o nome do aluno para remover: ")
    removido = False

    for aluno in alunos:
      if aluno["nome"] == nome:
        alunos.remove(aluno)
        removido = True
        print("Aluno removido com sucesso!")
        break

    if not removido:
      print("Aluno não encontrado.")

  case 7:
    # Listar_alunos()
    if len(alunos) == 0:
      print("Nenhum aluno cadastrado.")
    else:
      print("\nLista de alunos:")
      for aluno in alunos:
        print("-", aluno["nome"])

  case 0:
    print("Encerrando o sistema...")
    break

  case _:
    print("Opção inválida.")
```
